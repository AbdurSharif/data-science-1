**Task1**:

Describe the similarities and differences between an int and a float in python.

**Answer1**:

Similarities: Both represent numbers, both can be manipulated with operators, both are mutable.

Differences: ints are whole numbers, floats can be decimals.

**Task2**:

Describe the similarities and differences between: strings, lists, and dicts.

**Answer2**:

Similarities: All three can store written information, lists and dicts both seperate information in an arraylike form.

Differences: Strings don't have formal partitions between each piece of information, lists are based on indexes, dicts are based on keys. However, Strings are immutable, while lists and dicts are constantly changable, so when needing to change data a lot, its best to use lists or dicts. Additionally, when knowing how many categories you need to record, dicts can be more efficient than storing multiple lists.

**Task3**:

What is a numpy array? Why would we ever use one over a list or dict?

**Answer3**:

A numPy array is a list of one type of data, which can make it easier to use when just handling numbers, floats, etc. For example, 2D and 3D arrays or matrices can be used with numPy more efficiently and faster for complex calculations.

**Task4**:

What is a pandas dataframe? Why would we ever want to use one over a numpy array?

**Answer4**:

Its a 2D data structure that can be used for different types of variables, not just like numbers like in a numPy array.
